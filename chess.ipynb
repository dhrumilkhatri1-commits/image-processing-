```python
import numpy as np
import matplotlib.pyplot as plt

# User Input
board_size = int(input("Enter chessboard size (4, 8, etc.): "))
square_size = int(input("Enter square size (pixels): "))

color1 = input("Enter first color (black, white, red, green, blue): ").lower()
color2 = input("Enter second color (black, white, red, green, blue): ").lower()

# Color Dictionary
colors = {
    "black": [0, 0, 0],
    "white": [255, 255, 255],
    "red": [255, 0, 0],
    "green": [0, 255, 0],
    "blue": [0, 0, 255],
    "yellow": [255, 255, 0],
    "cyan": [0, 255, 255],
    "magenta": [255, 0, 255]
}

# Create Blank Image
chessboard = np.zeros((board_size * square_size,
                       board_size * square_size,
                       3), dtype=np.uint8)

for row in range(board_size):
    for col in range(board_size):

        if (row + col) % 2 == 0:
            chessboard[
                row*square_size:(row+1)*square_size,
                col*square_size:(col+1)*square_size
            ] = colors[color1]
        else:
            chessboard[
                row*square_size:(row+1)*square_size,
                col*square_size:(col+1)*square_size
            ] = colors[color2]

plt.imshow(chessboard)
plt.axis("off")
plt.show()
```
