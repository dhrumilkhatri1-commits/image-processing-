```python
# ==========================================
# Image Processing
# Image Cleaning + Image Manipulation
# ==========================================

import cv2
import numpy as np
import matplotlib.pyplot as plt
from skimage import data

# -----------------------------
# Load Sample Image
# -----------------------------
image = data.astronaut()   # Sample color image

plt.figure(figsize=(5,5))
plt.imshow(image)
plt.title("Original Image")
plt.axis("off")
plt.show()


# ==================================================
# Image Cleaning
# (Noise Removal using Gaussian Blur)
# ==================================================

# Add random noise
noise = np.random.normal(0, 25, image.shape).astype(np.uint8)
noisy_image = cv2.add(image, noise)

# Clean Image
clean_image = cv2.GaussianBlur(noisy_image, (5,5), 0)

plt.figure(figsize=(15,5))

plt.subplot(1,3,1)
plt.imshow(image)
plt.title("Original Image")
plt.axis("off")

plt.subplot(1,3,2)
plt.imshow(noisy_image)
plt.title("Noisy Image")
plt.axis("off")

plt.subplot(1,3,3)
plt.imshow(clean_image)
plt.title("Image Cleaning")
plt.axis("off")

plt.show()


# ==================================================
# Image Manipulation
# (Flip, Rotate, Resize)
# ==================================================

# Flip
flip_image = cv2.flip(image, 1)

# Rotate
(h, w) = image.shape[:2]
center = (w//2, h//2)

rotation_matrix = cv2.getRotationMatrix2D(center, 45, 1.0)
rotate_image = cv2.warpAffine(image, rotation_matrix, (w, h))

# Resize
resize_image = cv2.resize(image, (200,200))

plt.figure(figsize=(15,5))

plt.subplot(1,4,1)
plt.imshow(image)
plt.title("Original")
plt.axis("off")

plt.subplot(1,4,2)
plt.imshow(flip_image)
plt.title("Flip")
plt.axis("off")

plt.subplot(1,4,3)
plt.imshow(rotate_image)
plt.title("Rotate")
plt.axis("off")

plt.subplot(1,4,4)
plt.imshow(resize_image)
plt.title("Resize")
plt.axis("off")

plt.suptitle("Image Manipulation")
plt.show()
```
